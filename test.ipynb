```python
import pymysql
import pandas
import requests
import json
```

```python
url="https://data.moenv.gov.tw/api/v2/aqx_p_02?api_key=4c89a32a-a214-461b-bf29-30ff32a61a8a&limit=1000&sort=datacreationdate%20desc&format=JSON"
```

```python
res=requests.get(url)
res
```

```python
datas=res.json()
datas
```

```python
[i for i in datas]
```

```python
datas[0]
```

```python
data=[list(data.values()) for data in datas if list(data.values())[2]!=""]
data
```

```python
table_str="""
create table if not exists pm25(
id int auto_increment primary key,
site varchar(20),
county varchar(20),
pm25 int,
datacreationdate datetime,
itemunit varchar(20),
unique key site_time (site,datacreationdate)
)
"""
```

```python
conn=pymysql.connect(
    host="localhost",
    password="",
    port=3306,
    user="root",
    database="demo",
)
conn
cursor=conn.cursor()
cursor
```

```python
cursor.execute(table_str)
conn.commit()
```

```python
conn.close()
```

```python
str_sql="""
insert into pm25(site,county,pm25,datacreationdate,itemunit) values(%s,%s,%s,%s,%s)
"""
```

```python
cursor.executemany(str_sql,data)
conn.commit()
```
